### Building simple linear model

```python
def compute_model_output(x,w,b):
    m=x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i]=w*x[i]+b
    return f_wb
```

```python
temp_ans = compute_model_output(x_train,100,100)
```

```python
temp_ans
```

## plotting model prediction

```python
plt.plot(x_train,temp_ans)
plt.scatter(x_train,y_train)
```

```python
temp_ans = compute_model_output(x_train,50,100)
temp_ans
```

```python
plt.plot(x_train,temp_ans)
plt.scatter(x_train,y_train)
```
