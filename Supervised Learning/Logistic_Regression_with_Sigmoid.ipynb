### Sigmoid function

```python
x_train = np.arange(-10,10,2)
```

```python
x_train
```

```python
def sigmoid(z):
    g = 1/(1+(1/np.exp(-z)))
    return g
```

```python
print(sigmoid(0))
```

```python
y_train=sigmoid(x_train)
```

```python
y_train
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
plt.plot(x_train,y_train)
```

```python
# Plot z vs sigmoid(z)
fig,ax = plt.subplots(1,1,figsize=(5,3))
ax.plot(x_train, y_train, c="b")

ax.set_title("Sigmoid function")
ax.set_ylabel('sigmoid(z)')
ax.set_xlabel('z')
```
